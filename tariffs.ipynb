```python
from datetime import datetime

by_year = [
    df_2013,
    df_2014,
    df_2015,
    df_2016,
    df_2017,
    df_2018,
    df_2019,
    df_2020
]

for i in range(len(by_year)):
    by_year[i].insert(0, "Year", datetime(i + 2013, 1, 1))
    
by_year[0]
```

```python
df = pd.concat(by_year)
df.set_index("Year", inplace=True)

print(df.shape)
df.head()
```

```python
by_country = {}

for country in list(df.Country.value_counts().keys()):
    if df.Country.value_counts()[country] == 8:
        by_country[country] = df[df.Country == country]
        
len(by_country.keys())
```

```python
for country in by_country.keys():
    print(country)
```

```python
avg_tariff = []

for country in by_country.keys():
    avg_tariff.append((country, by_country[country]["Tariff Rate (%)"][:"2017-01-01"].mean()))
    
len(avg_tariff)
```

```python
max_tariff = []

for country in by_country.keys():
    max_tariff.append((country, by_country[country]["Tariff Rate (%)"][:"2017-01-01"].max()))
    
len(max_tariff)
```

```python
max_tariff
```

```python
min_tariff = []

for country in by_country.keys():
    min_tariff.append((country, by_country[country]["Tariff Rate (%)"][:"2017-01-01"].min()))
    
len(min_tariff)
```

```python
min_tariff
```

```python
tariff_rate = {}

for i in range(len(avg_tariff)):
    tariff_rate[avg_tariff[i][0]] = (min_tariff[i][1], avg_tariff[i][1], max_tariff[i][1])
    
len(tariff_rate.keys())
```

```python
data = {}

for i in range(len(tariff_rate.keys())):
    data[list(tariff_rate.keys())[i]] = {}
    
for i in range(len(data.keys())):
    data[list(data.keys())[i]]["Minimum"] = min_tariff[i][1]
    data[list(data.keys())[i]]["Average"] = avg_tariff[i][1]
    data[list(data.keys())[i]]["Maximum"] = max_tariff[i][1]
    
len(data.keys())
```

```python
data["Kiribati"]
```

```python
tariffs = pd.DataFrame.from_dict(data).T

print(tariffs.shape)
tariffs.head()
```
