```python
from datetime import datetime

by_year = [
    df_2013,
    df_2014,
    df_2015,
    df_2016,
    df_2017,
    df_2018,
    df_2019,
    df_2020
]

for i in range(len(by_year)):
    by_year[i].insert(0, "Year", datetime(i + 2013, 1, 1))
    
by_year[0]
```

```python
df = pd.concat(by_year)

print(df.shape)
df.head()
```

```python
by_country = {}

for country in list(df.Country.value_counts().keys()):
    if df.Country.value_counts()[country] == 8:
        by_country[country] = df[df.Country == country]
        
len(by_country.keys())
```

```python
for country in by_country.keys():
    print(country)
```

```python
avg_tariff = []

for country in by_country.keys():
    avg_tariff.append((country, by_country[country]["Tariff Rate (%)"].mean()))
    
len(avg_tariff)
```

```python
avg_tariff.sort(key=lambda x: x[1], reverse=True)
avg_tariff
```
